The $\alpha$ â€“ asynchronous cellular automata is a more general case, where each cell undergo the local transformation with probability $\alpha$. This is more realistic to nature.  


```python
import random

#This lambda function create an arbitrary binary strings of length n

random_binary_string = lambda n: ''.join(random.choice('01') for _ in range(n)) 

# this lambda function is a helper function that convert the rule number in binary format of 8 characters
binary_r = lambda r: format(r, f'0{8}b')

def alphaAsynchronus(alpha,rule,time,n_length):

    automata_rule = {}
    r = binary_r(rule)  
    container_bin = [format(i, f'0{3}b') for i in range(2**3)][::-1]    #output:['111', '110', '101', '100', '011', '010', '001', '000']
    for i,key in enumerate(container_bin):
        automata_rule[key] = r[i]
    current = random_binary_string(n_length)
    print(current)
    while time>0:
        
        exec = current[-1] + current + current[0]
        new = ''
        for i in range(len(current)):                        # this piece of code ensure the effect of alpha; the cell updates 
            if random.random()<alpha:                        # itself only when a random number between 0 and 1 is lesser the alpha,
                new += automata_rule[exec[i:i+3]]            # thus on average the ratio of updated cell to total cell is alpha                                     
            else:                                                
                new += current[i]
        current = new
        time -= 1
        print(current)
```

```python
latticeNumber = 10
# alpha = 0.8 ; rule = 30 ; timestamp = 20
alphaAsynchronus(0.8,30,20,latticeNumber)
```
